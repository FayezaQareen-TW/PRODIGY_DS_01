```python
import pandas as pd
import matplotlib.pyplot as pltR
import seaborn as sns

# 1. Load the data (skipping the first 4 metadata rows)
# Note: If your file name ends slightly differently, change it below
filename = 'API_SP.POP.TOTL_DS2_en_csv_v2_38144 (1).csv'

try:
    df = pd.read_csv(filename, skiprows=4)
except FileNotFoundError:
    # Fallback if the name uploaded without the " (1)" suffix
    df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)

# 2. Drop rows where the 2020 population data is missing
df = df.dropna(subset=['2020'])

# 3. Select a small group of countries to compare
sample_countries = ['India', 'United States', 'Indonesia', 'Brazil', 'Pakistan']
filtered_df = df[df['Country Name'].isin(sample_countries)]

# 4. Create a clean bar chart for the year 2020
plt.figure(figsize=(10, 6))
sns.barplot(x='Country Name', y='2020', data=filtered_df, palette='muted')
plt.title('Total Population Comparison in 2020 (Prodigy InfoTech Task 1)')
plt.ylabel('Population (in Billions)')
plt.xlabel('Country Name')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('prodigy_task_1_chart.png',dpi=300,bbox_inches='tight')
plt.show()
```
